```python
df_dropped = df_FENZ.drop(['Second_Arrival_Time','Flame_Damage','TLA','Address_No_','Street','Street_Type','Suburb','Incident_Type','Date_Time','Percent_of_Property_Saved_Name','Ethnicity','Age_of_Person_Involved','Location_of_Origin','Origin_Group','Fire_Cause','Fire_Cause_Group','Heat_Source','Heat_Source_Group','Construction_Type','Year_Constructed',
             'Specific_Property_Use','Specific_Property_Use_Group','Specific_Property_Use_Group_Parent','General_Property_Use','General_Property_Use_Group'],axis=1)
df_dropped = df_dropped.dropna()
# Correct time of day format to be average times
df_dropped['Time_of_Day'] = [int(x.split(' - ')[0].split(':')[0])*60+30 for x in df_dropped.Time_of_Day]

df_dropped['First_Arrival_Time'] = [x.split(' ') for x in df_dropped['First_Arrival_Time']]

df_dropped['FirstArrivalTime'] = [int(x[0]) if len(x)==2 else 0 for x in df_dropped['First_Arrival_Time']]
df_dropped['FirstArrivalTime'] = [y+float(x[0])+float(x[2])/60 if len(x)==4 else y for x,y in zip(df_dropped['First_Arrival_Time'],df_dropped['FirstArrivalTime'])]
df_dropped['FirstArrivalTime'] = [y+float(x[0])*60+float(x[2])+float(x[4])/60 if (len(x)==6) else y for x,y in zip(df_dropped['First_Arrival_Time'],df_dropped['FirstArrivalTime'])]
df_dropped['FirstArrivalTime'] = [y+float(x[0])*24*60+float(x[2])*60+float(x[4])+float(x[6])/60 if (len(x)==8) else y for x,y in zip(df_dropped['First_Arrival_Time'],df_dropped['FirstArrivalTime'])]

df_dropped.drop('First_Arrival_Time',axis=1,inplace=True)

df_modelling = df_dropped.drop(['Incident_ID','Exposure_ID','MapgridE__X_','MapgridN__Y_','Unclassified_Injury_Count'],axis=1)
df_modelling = df_modelling[df_modelling['Numfloors']!=3150]
df_modelling = df_modelling[df_modelling['Structure_Area__m2_']!=999990]
df_modelling = df_dropped.drop(['Incident_ID','Exposure_ID','MapgridE__X_','MapgridN__Y_','Unclassified_Injury_Count'],axis=1)
```

```python
casualty_list = ['Casualty_Count','Fatality_Count','Critical_Injury_Count','Serious_Injury_Count','Moderate_Injury_Count','Minor_Injury_Count']
for x in list(df_modelling):
    if type(df_modelling[x][0])!=str:
        if x not in ['Percent_of_Property_Saved','Numfloors','Floor_Area__m2_','Flame_Damage__m2_','Structure_Area__m2_','Casualty_Count','Fatality_Count',
                    'Critical_Injury_Count','Serious_Injury_Count','Moderate_Injury_Count','Minor_Injury_Count']:
            bins = [df_modelling[x].quantile(0),df_modelling[x].quantile(0.33),df_modelling[x].quantile(0.67),df_modelling[x].quantile(1)]
        elif x=='Percent_of_Property_Saved':
            bins = [0,15,19,23]
        elif x=='Numfloors':
            bins = [0,1.5,5.5,130]
        elif x=='Floor_Area__m2_':
            bins = [0,100,1000,100000]
        elif (x=='Flame_Damage__m2_')|(x=='Structure_Area__m2_'):
            bins = [0,20,200,100000]
        elif x in casualty_list:
            bins = [-1,0.1,100]
        
        if x not in casualty_list:
            df_modelling[x] = pd.cut(df_modelling[x],bins=bins,labels=['Low','Medium','High'])
        else:
            df_modelling[x] = pd.cut(df_modelling[x],bins=bins,labels=['Zero','AtLeastOne'])
```

```python
df_modelling = df_modelling.dropna()
```

```python
df_modelling.columns = [x.replace('_','') for x in df_modelling]
```

```python
df_modelling.drop(['SpecificPropertyUseName','SpecificPropertyUseGroupName','LocationofOriginName','FireCauseName','GeneralPropertyUseName','HeatSourceName'],axis=1,inplace=True)
```

```python
df_modelling = pd.DataFrame({x:df_modelling[x].str.replace(',','').str.replace(' ','').str.replace('-','').str.replace(':','').str.replace(' ','') for x in list(df_modelling)})
```

```python
df_modelling.to_csv('dataset.csv',index=False)
```

```python
net = pysmile.Network()

for col in list(df_modelling):
    net.add_node(pysmile.NodeType.CPT,col)
    
    for outcome in np.unique(list(df_modelling[col])):
        net.add_outcome(col,outcome)
        
    for outcome in ['State0','State1']:
        net.delete_outcome(col,outcome)
```

```python
ds = pysmile.learning.DataSet()
ds.read_file("dataset.csv")
bayes_search = pysmile.learning.BayesianSearch()
net = bayes_search.learn(ds)
```
